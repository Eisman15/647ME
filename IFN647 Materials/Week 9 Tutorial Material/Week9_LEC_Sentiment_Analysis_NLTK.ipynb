```python
# import libraries
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# Load the amazon review dataset

df = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
#df = pd.read_csv('amazon.csv')
df
```

```python
# Define a function parse_text function for text pre-processing
def parse_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens (like stemming but it brings context to the words, and links words with similar meanings to one word)
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text
```

```python
# apply the parse_text function via df
df['reviewText'] = df['reviewText'].apply(parse_text)
df
```

```python
# initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Define cal_sentiment_scor() function, which uses polarity_scores() to obtain a dictionary of 
#  sentiment scores ([-1, +1]) for positive, negative or neutral sentiments.
# The polarity is then determined by checking whether the positive score is greater than a threshold theta,  
#  then returns a sentiment score of 1 if it is, and a 0 otherwise (a binary decison making). 

def cal_sentiment_score(text):
    theta = 0.15 # a threshold 
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > theta else 0
    return sentiment

# apply get_sentiment function
df['sentiment'] = df['reviewText'].apply(cal_sentiment_score)

df
```

```python
# Evaluate the proposed get_sentiment() function by comparing two columns 'Positive' and 'sentiment' 
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['Positive'], df['sentiment']))

from sklearn.metrics import classification_report
print(classification_report(df['Positive'], df['sentiment']))
```
