```python
import nltk
# import libraries
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# (a)	Load the dataset of Amazon reviews using pd.read_csv() and define a function parse_text() to preprocess the reviewText column
```

```python
# Load the amazon review dataset
df = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
```

```python
# Define a function parse_text function for text pre-processing
def parse_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens (like stemming but it brings context to the words, and links words with similar meanings to one word)
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text
```

```python
# apply the parse_text function via df
df['reviewText'] = df['reviewText'].apply(parse_text)
```

```python
# (b) Convert the reviewText column into a list of Python strings, docs, and the positive column into a list of Python integers, y_true
```

```python
# view each reviewText as a document and store them in a list
docs = df['reviewText'].astype(str).tolist()
```

```python
# get the ground truth labels
y_true = df['Positive'].astype(int).tolist()
```

```python
# (c) Design an LDA-based clustering model using LatentDirichletAllocation and CountVectorizer
```

```python
#  LDA (model)-based clustering
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)

lda = LatentDirichletAllocation(n_components=3)
# n_components means number of topics
lda.fit(X)
```

```python
# (d)	Get document-topic distribution and print top-5 words for the three topics.
```

```python
# for each document, it finds the topics distribution in the document 
# Get document-topic distribution
doc_topics = lda.transform(X)
print("\nDocument-topic distribution:")
print(doc_topics)
```

```python
# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Print top words per topic
n_top_words = 5
for topic_idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[-n_top_words:][::-1]]
    print(f"Topic {topic_idx}: {top_words}")
```

```python
# Assign clusters (hard clustering)
clusters = doc_topics.argmax(axis=1)
print("\nDocument clusters:")
y_pred = []
for i, cluster in enumerate(clusters):
    print(f"Doc {i} -> Cluster {cluster}")
    y_pred.append(cluster)
```

```python
# Purity Calculation Code
print(y_pred)
```

```python
# (f)	Design a purity score function to evaluate the clustering performance.
```

```python
import numpy as np
from collections import Counter

def purity_score(y_true, y_pred):
    """
    Compute clustering purity

    Parameters:
    y_true : list or array of true labels
    y_pred : list or array of cluster assignments

    Returns:
    purity : float
    """
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    total_correct = 0

    for cluster in np.unique(y_pred):
        # indices of points in this cluster
        indices = np.where(y_pred == cluster)[0]
        #print(indices)
        # true labels of these points
        labels = y_true[indices]
        #print(labels)
        # count most common label
        most_common_count = Counter(labels).most_common(1)[0][1]
        #print(most_common_count)
        total_correct += most_common_count

    return total_correct / len(y_true)
```
